# Encoding qilish

```python
# turli xil lenga ega shuning uchun uni encoding qilish kerak
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))  # Initialize a matrix of zeros

    # Loop over each sequence
    for i, sequence in enumerate(sequences):
        # Set the corresponding indices to 1
        for j in sequence:
            results[i, j] = 1.
    return results


```

```python
# Vectorize  training and testing data
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
#Endi data tayyor training qilish uchun
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")
```

# Model Building

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Assert that x_train and y_train have matching sizes
assert x_train.shape[0] == y_train.shape[0], "Mismatch between x_train and y_train sizes"


# Split data into training and validation sets
x_val = x_train[:10000]
partial_x_train = x_train[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]


```

```python
# Define the model
model = keras.Sequential([
    layers.Dense(16, activation="relu"),  # 16-dimensional representations
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])


```

```python
# Train the model
history = model.fit(
    partial_x_train, y=partial_y_train,
    epochs=20,
    batch_size=512,
    validation_data=(x_val, y_val)  # Validation data
)

# Evaluate the model on the test data
results = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {results[1]}")
```
